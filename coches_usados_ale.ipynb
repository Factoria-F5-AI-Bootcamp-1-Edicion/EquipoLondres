# **Preparación**

```python
# "drop" elimina columnas.
df.drop(["Unnamed: 0", "index", "carID"], axis=1, inplace=True)
df.head()
```

**Variables:**
+ brand = marca del coche.
+ model = modelo del coche.
+ year = año.
+ transmission = automático, manual, semi-auto (...).
+ mileage = kilometraje.
+ fuelType = tipo de combustible.
+ tax = impuesto.
+ mpg = (?)
+ engineSize = tamaño del motor.
+ price = precio.

```python
#variables a predecir: price
# variable número -> modelo de regresión
X = df.drop ("price", axis=1)
y = df ["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
categoricas = ["brand", "model", "transmission", "fuelType"]
numericas = ["year", "mileage", "tax", "mpg", "engineSize"]
```

```python
#MinmaxScaler para numericas y OneHotEncoder para categoricas.
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer_numerico = ("transformer_numerico", MinMaxScaler(), numericas)
transformer_categorico = ("transformer_categorico", OneHotEncoder(), categoricas)

transformer = ColumnTransformer([transformer_numerico, transformer_categorico], remainder="passthrough")
```

```python
#Vector disperso a "array"
X = transformer.fit_transform(X).toarray()
```

```python
#Vector disperso: nuevas columnas "MinmaxScaler para numericas y OneHotEncoder para categoricas". Objetivo = numéricas.
pd.DataFrame(X, columns = transformer.get_feature_names_out())
```

```python
transformer.output_indices_
```

# **Entrenamos**

```python
#Voy a entrenar sobre "X y Y train", sobre "X y Y test".
#Hasta ahora docu Sk-learn
# X en mayúscula (convención) = matriz, y = vector.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y) #Aleatorio
```

```python
# Importar códigos de los modelos (algoritmos) de la docu de sklearn 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

#métricas para saber si está bien o mal mi modelo
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#creando función: instancio el modelo, ajusto a mis datos de entrenamiento, predicción sobre los datos que he apartado para el test
def train_evaluate(nombre_modelo, modelo):
    mod = modelo()
    mod.fit(X_train, y_train)
    
    y_predict = mod.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_predict)
    mse = mean_squared_error(y_test, y_predict)
    rmse = mean_squared_error(y_test, y_predict, squared=False) #raíz cuadrada
    r2 = r2_score(y_test, y_predict)
    
    print(nombre_modelo)
    print()
    print(f"MAE: {mae}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"R2: {r2}")
    print()
```

```python
# Errores en modelo de regresión lineal
train_evaluate("LinearRegression", LinearRegression)
```

```python
# Errores en modelo SVM
train_evaluate("SVM",SVR)
```

```python
# Errores en modelo Árbol de Decisión
train_evaluate("Decission Tree", DecisionTreeRegressor)
```
